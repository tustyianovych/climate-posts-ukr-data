The distribution you're describing, which decreases from 0.0 to ~0.5 and then increases again towards 1.0, with both ends (0.0 and 1.0) being nearly at the same level, is typically referred to as a **U-shaped distribution**. This type of distribution is characterized by higher frequencies at the extremes and a lower frequency in the middle range. It suggests that the values are more commonly found at the low and high ends of the scale, with fewer occurrences around the midpoint.

#### Compute a score between -1 and 1 to depict reactions from negative, neutral to positive

```python
def calculate_reactions_score(row):
    positive_percent = row['Positive_Percent'] * 1
    negative_percent = row['Negative_Percent'] * -1
    neutral_percent = row['Neutral_Percent'] * 0

    total_score = positive_percent + negative_percent + neutral_percent

    return total_score
```

```python
reactions_df['reactions_score'] = reactions_df.apply(lambda x: calculate_reactions_score(x), axis=1)
```

```python
sns.histplot(data=reactions_df, x="reactions_score", bins=50, kde=True)
plt.show()
```

```python
def preprocess(message):
    message = message.str.replace("(<br/>)", "")
    message = message.str.replace('(<a).*(>).*(</a>)', '')
    message = message.str.replace('(&amp)', '')
    message = message.str.replace('(&gt)', '')
    message = message.str.replace('(&lt)', '')
    message = message.str.replace('(\xa0)', ' ')  
    return message
    
posts_reactions_df['processed_text'] = preprocess(posts_reactions_df['processed_text'])

posts_reactions_df['text_len'] = posts_reactions_df['processed_text'].astype(str).apply(len)
posts_reactions_df['word_count'] = posts_reactions_df['processed_text'].apply(lambda x: len(str(x).split()))
```

```python
from sklearn.feature_extraction.text import CountVectorizer


def get_top_n_words(corpus, n=None):
    vec = CountVectorizer().fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
    
common_words = get_top_n_words(posts_reactions_df['text_preprocessed'], 20)
for word, freq in common_words:
    print(word, freq)
df1 = pd.DataFrame(common_words, columns = ['text' , 'count'])
sns.barplot(data = df1.groupby('text').sum()['count'].sort_values(ascending=False))
# df1.groupby('text').sum()['count'].sort_values(ascending=False).iplot(
#     kind='bar', yTitle='Count', linecolor='black', title='Top 20 words in the climate posts before removing stop words')
```

```python
def token_filter(token, stopwords):
    return not (
        token.is_punct |
        token.is_space |
        token.is_stop |
        len(token.text) <= 2 |
        (token in stopwords)
        )


def spacy_parallel_doc_processing(nlp, stopwords, docs):
    print('spacy_parallel_doc_processing called')

    filtered_tokens = []

    for index, doc in enumerate(nlp.pipe(docs, n_process=-1)):

        tokens = [token.lemma_ for token in doc
                  if token_filter(token, stopwords)]
        tokens_filtered = [token for token in tokens if token not in stopwords]
        doc_result = ' '.join(tokens_filtered)

        filtered_tokens.append(doc_result)

    return filtered_tokensb


def spacy_doc_processing(nlp, stopwords, docs):
    print('spacy_doc_processing called')

    filtered_tokens = []

    for index, doc in enumerate(docs):
        try:
            nlp_doc = nlp(doc)
            tokens = [token.lemma_ for token in nlp_doc
                      if token_filter(token, stopwords)]
            tokens_filtered = [token for token in tokens
                               if token not in stopwords]
            doc_result = ' '.join(tokens_filtered)
    
            filtered_tokens.append(doc_result)

        except Exception as e:
            print(e)
            print(index, doc)
            break

    return filtered_tokens
```

### Spacy and ScatterText to Analyze how different groups of text sentiment differ

```python
import pandas as pd
import scattertext as st
import spacy

nlp = spacy.load("uk_core_news_lg")
```

```python
def reaction_category(reaction_score):
    if reaction_score <= -0.5:
        return "Strong negative"
    elif reaction_score > -0.5 and reaction_score < 0:
        return "Negative"
    elif reaction_score > 0 and reaction_score < 0.5:
        return "Positive"
    elif reaction_score >= 0.5:
        return "Strong positive"
    elif reaction_score == 0:
        return "Neutral"
    else:
        return "Unavailable"
```

### Calculate N-grams on lemmatized text

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
def get_top_n_words(corpus, n=None):
    vec = CountVectorizer(stop_words = ua_stopwords).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
    
common_words = get_top_n_words(posts_reactions_df['lemmatized_text'], 30)
for word, freq in common_words:
    print(word, freq)
lemmatized_text_unigram_df = pd.DataFrame(common_words, columns = ['text' , 'count'])
```

```python
def get_top_n_bigram(corpus, n=None):
    vec = CountVectorizer(ngram_range=(2, 2), stop_words = ua_stopwords).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
    
common_words = get_top_n_bigram(posts_reactions_df['lemmatized_text'], 30)
for word, freq in common_words:
    print(word, freq)
lemmatized_text_bigram_df = pd.DataFrame(common_words, columns = ['text' , 'count'])
```

```python
def get_top_n_trigram(corpus, n=None):
    vec = CountVectorizer(ngram_range=(3, 3), stop_words = ua_stopwords).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
    
common_words = get_top_n_trigram(posts_reactions_df['lemmatized_text'], 30)
for word, freq in common_words:
    print(word, freq)
lemmatized_text_trigram_df = pd.DataFrame(common_words, columns = ['text' , 'count'])
```

```python
def get_top_n_fourgram(corpus, n=None):
    vec = CountVectorizer(ngram_range=(4, 4), stop_words = ua_stopwords).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
    
common_words = get_top_n_fourgram(posts_reactions_df['lemmatized_text'], 30)
for word, freq in common_words:
    print(word, freq)
lemmatized_text_fourgram_df = pd.DataFrame(common_words, columns = ['text' , 'count'])
```

```python
lemmatized_ngrams_df = pd.concat([lemmatized_text_trigram_df, lemmatized_text_bigram_df,
                                 lemmatized_text_unigram_df, lemmatized_text_fourgram_df])
```
