```python
import pandas as pd
import re

f = open("structural equations modeling results.txt")
sem_results = f.read()
```

```python
# sem_results_split = re.split('\b[A-Z\s]+ MODEL', sem_results)
```

```python
sem_results_split = re.split('\n\n', sem_results)
```

```python
total_rows = 0
structural_equations_modeling_consolidated = []


for model_output in sem_results_split:
    # model_name = re.search('\b[A-Z\s]+ MODEL', model_output)
    model_name = re.search(r'(\b[A-Z\s]+ MODEL)', model_output).group(0)
    model_output_split = model_output.split('\n')
    total_rows += len(model_output_split) - 1
    for model_output_row in model_output_split[1:]:
        output_entity = {
            'model_name': model_name
        }
        extracted_entities = re.search(r'([\w_]+)[\s~]+([\w_]+)[\s=]+([\d.-]+)\s\(SE: ([\d.-]+), p: ([\d.-]+)\)([\s\*]+)\[(\w+)\]', model_output_row)
        output_entity['target_variable'] = extracted_entities.group(1)
        output_entity['predictor'] = extracted_entities.group(2)
        output_entity['coef'] = extracted_entities.group(3)
        output_entity['st.err'] = extracted_entities.group(4)
        output_entity['p_value'] = extracted_entities.group(5)
        output_entity['significance'] = extracted_entities.group(6)
        output_entity['effect_size'] = extracted_entities.group(7)

        structural_equations_modeling_consolidated.append(output_entity)
        
        
    
```

```python
print(total_rows)
```

```python
len(structural_equations_modeling_consolidated)
```

```python
structural_equations_modeling_consolidated_df = pd.DataFrame(structural_equations_modeling_consolidated)
```

```python
structural_equations_modeling_consolidated_df.head()
```

```python
structural_equations_modeling_consolidated_df['p_value'] = pd.to_numeric(structural_equations_modeling_consolidated_df['p_value'])
structural_equations_modeling_consolidated_df['coef'] = pd.to_numeric(structural_equations_modeling_consolidated_df['coef'])
structural_equations_modeling_consolidated_df['abs_coef'] = structural_equations_modeling_consolidated_df['coef'].abs()
structural_equations_modeling_consolidated_df['st.err'] = pd.to_numeric(structural_equations_modeling_consolidated_df['st.err'])
structural_equations_modeling_consolidated_df['coef_ratio'] = structural_equations_modeling_consolidated_df['coef'] / structural_equations_modeling_consolidated_df['st.err']
structural_equations_modeling_consolidated_df['abs_coef_ratio'] = structural_equations_modeling_consolidated_df['coef_ratio'].abs()
```

```python
structural_equations_modeling_consolidated_df
```

```python
strong_relationships_df = structural_equations_modeling_consolidated_df[structural_equations_modeling_consolidated_df['abs_coef_ratio'] > 2]
weak_relationships_df = structural_equations_modeling_consolidated_df[structural_equations_modeling_consolidated_df['abs_coef_ratio'] <= 2]
```
