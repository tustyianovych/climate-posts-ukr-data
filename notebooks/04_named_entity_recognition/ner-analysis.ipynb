```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_parquet("data/all_collected_text_data_processed.parquet")
```

```python
ner_results = []

for index, items in enumerate(df.named_entities_raw.to_list()):
    named_entities = items.split(', ')
    for entity in named_entities:
        entity_info = {
            "id": index,
            "value": entity.replace('\u200b', '')
        }

        ner_results.append(entity_info)
```

# Association rules per topic

```python
import pandas as pd
from IPython.display import display, HTML
from pyvis.network import Network

from mlxtend.preprocessing import TransactionEncoder
```

```python
def generate_network_graph(name, dataset):

    got_net = Network(
        notebook=True,
        cdn_resources="remote",
        height="500px",
        width="100%",
        bgcolor="white",
        font_color="red",
    )
    
    # set the physics layout of the network
    got_net.repulsion()
    got_data = df
    
    sources = dataset.antecedents_str
    targets = dataset.consequents_str
    weights = dataset.lift
    
    edge_data = zip(sources, targets, weights)
    
    for e in edge_data:
        src = e[0]
        dst = e[1]
        w = e[2]
    
        got_net.add_node(src, src, title=src)
        got_net.add_node(dst, dst, title=dst)
        got_net.add_edge(src, dst, value=w)
    
    neighbor_map = got_net.get_adj_list()
    
    got_net.show(f"ner-association-rules/{name}.html")
```

```python
def generate_association_rules(dataframe, topic):
    topic_df = dataframe[dataframe['Primary Topic'] == topic]

    flair_ukr_ner_named_entities = topic_df.flair_ukr_ner_named_entities.to_list()
    spacy_ukr_ner_named_entities = topic_df.named_entities_raw.to_list()
    
    flair_ukr_ner_named_entities_list = []
    spacy_ukr_ner_named_entities_list = []


    for index, items in enumerate(spacy_ukr_ner_named_entities):
        named_entities = [item.replace('\u200b', '') for item in items.split(', ')]
        spacy_ukr_ner_named_entities_list.append(named_entities)
    
    for items in flair_ukr_ner_named_entities:
        flair_ukr_ner_named_entities_list.append(list(items))

    flair_te = TransactionEncoder()
    spacy_te = TransactionEncoder()

    flair_oht_ary = flair_te.fit(flair_ukr_ner_named_entities_list).transform(flair_ukr_ner_named_entities_list, sparse=True)
    flair_ukr_ner_named_entities_sparse_df = pd.DataFrame.sparse.from_spmatrix(flair_oht_ary, columns=flair_te.columns_)

    spacy_oht_ary = spacy_te.fit(spacy_ukr_ner_named_entities_list).transform(spacy_ukr_ner_named_entities_list, sparse=True)
    spacy_ukr_ner_named_entities_sparse_df = pd.DataFrame.sparse.from_spmatrix(spacy_oht_ary, columns=spacy_te.columns_)

    flair_ner_frequent_itemsets_sparse = apriori(flair_ukr_ner_named_entities_sparse_df, min_support=0.0025, use_colnames=True, verbose=1, max_len=3, low_memory=False)
    spacy_ner_frequent_itemsets_sparse = apriori(spacy_ukr_ner_named_entities_sparse_df, min_support=0.0025, use_colnames=True, verbose=1, max_len=3, low_memory=False)

    flair_climate_rules = association_rules(flair_ner_frequent_itemsets_sparse, metric ="confidence") 
    flair_climate_rules = flair_climate_rules.sort_values(['confidence', 'lift'], ascending =[False, False]) 

    spacy_climate_rules = association_rules(spacy_ner_frequent_itemsets_sparse, metric ="confidence") 
    spacy_climate_rules = spacy_climate_rules.sort_values(['confidence', 'lift'], ascending =[False, False])

    spacy_climate_rules['antecedents_length'] = spacy_climate_rules.antecedents.apply(len)
    spacy_climate_rules['consequents_length'] = spacy_climate_rules.consequents.apply(len)

    flair_climate_rules['antecedents_length'] = flair_climate_rules.antecedents.apply(len)
    flair_climate_rules['consequents_length'] = flair_climate_rules.consequents.apply(len)
    
    flair_single_item_climate_rules = flair_climate_rules[(flair_climate_rules.antecedents_length == 1) & (flair_climate_rules.consequents_length == 1)]
    spacy_single_item_climate_rules = spacy_climate_rules[(spacy_climate_rules.antecedents_length == 1) & (spacy_climate_rules.consequents_length == 1)]
    
    flair_single_item_climate_rules["antecedents_str"] = flair_single_item_climate_rules["antecedents"].apply(lambda x: list(x)[0]).astype("unicode")
    flair_single_item_climate_rules["consequents_str"] = flair_single_item_climate_rules["consequents"].apply(lambda x: list(x)[0]).astype("unicode")

    spacy_single_item_climate_rules["antecedents_str"] = spacy_single_item_climate_rules["antecedents"].apply(lambda x: list(x)[0]).astype("unicode")
    spacy_single_item_climate_rules["consequents_str"] = spacy_single_item_climate_rules["consequents"].apply(lambda x: list(x)[0]).astype("unicode")

    topic_name_formatted = topic.lower().replace(" ", "_").replace("-", "_")

    flair_network_graph_name = "flair_{0}".format(topic_name_formatted)
    spacy_network_graph_name = "spacy_{0}".format(topic_name_formatted)

    generate_network_graph(flair_network_graph_name, flair_single_item_climate_rules)
    generate_network_graph(spacy_network_graph_name, spacy_single_item_climate_rules)
```

```python
df['Primary Topic'].unique()
primary_topics = ["Energy and Resources"]
```

```python
for topic in primary_topics:
    generate_association_rules(df, topic)
```
