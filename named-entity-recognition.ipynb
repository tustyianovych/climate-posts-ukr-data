```python
def token_filter(token, stopwords):
    return not (
        token.is_punct |
        token.is_space |
        token.is_stop |
        len(token.text) <= 2 |
        (token in stopwords)
        )
```

```python
def spacy_doc_processing(nlp, stopwords, docs):
    print('spacy_doc_processing called')

    filtered_tokens = []

    for index, doc in enumerate(docs):
        try:
            nlp_doc = nlp(doc)
            tokens = [token.lemma_ for token in nlp_doc
                      if token_filter(token, stopwords)]
            tokens_filtered = [token for token in tokens
                               if token not in stopwords]
            doc_result = ' '.join(tokens_filtered)
    
            filtered_tokens.append(doc_result)

        except Exception as e:
            print(e)
            print(index, doc)
            break

    return filtered_tokens
```

```python
ua_stopwords = pd.read_csv("utils/ua-stopwords.txt")

spacy_ukr_stopwords = nlp.Defaults.stop_words

for stopword in ua_stopwords:
    spacy_ukr_stopwords.add(stopword)
```

```python
cleaned_text = spacy_doc_processing(nlp, spacy_ukr_stopwords, docs)
```

```python
corpus_raw = df['text'].to_list()
```

```python
start_time = time.time()


flair_ukr_ner_named_entities = []

for index, item in enumerate(corpus_raw):
    try:
        
        extracted_named_entities = []
        
        sentence = Sentence(item)
        tagger.predict(sentence)

        entities = sentence.get_labels()
        for entity in entities:
            extracted_named_entities.append(entity.shortstring)
        
        flair_ukr_ner_named_entities.append(extracted_named_entities)
            
    except Exception as e:
        print(e)
        print(index, item)


print("--- %s seconds ---" % (time.time() - start_time))
```
